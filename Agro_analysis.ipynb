```python
def fill_0(df):
    df_mean=df.mean(axis=1).values
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if df.iloc[i,j]==0:
                df.iloc[i,j]=df_mean[i]
    return df
```
